```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
x = np.linspace(0, 2*np.pi, 1000)

a1=1
a2=3
a3=5

s1 = np.sin(a1*x)
s2 = np.sin(a2*x)
s3 = np.sin(a3*x)


plt.plot(x, s1, label='sin(a1)', color='red')
plt.plot(x, s2, label='sin(a2)', color='green')
plt.plot(x, s3, label='sin(a3)', color='blue')


plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('3 Differencet Sine waves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
amplitude_factor = 2

s1 = amplitude_factor * np.sin(a1*x)
s2 = amplitude_factor * np.sin(a2*x)
s3 = amplitude_factor * np.sin(a3*x)

a4 = 7
a5 = 2
phi1 = np.pi/4

s4 = amplitude_factor * np.sin(a4 * x + phi1)
s5 = amplitude_factor * np.sin(a5 * x)


plt.plot(x, s1, label='sin(a1*x)', color='red')
plt.plot(x, s5, label=f'sin({a5}*x)', color='orange')

plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Waves with Increased Amplitudes and New Waves')
plt.legend()
plt.grid(True)
plt.show()
```
